```python
df['new_title'] = df['Title'].str.split(': Season').str[0]
df['season'] = df['Title'].str.extract(r'Season (\d+)')
df = df[df['season'].isna() == False].sort_values(['new_title','season'])#.sort_values('season')
df = df[['new_title', 'season', 'Hours Viewed']]
df
```

```python
# Assuming your DataFrame is named df
# Convert 'Hours Viewed' column to numeric
df['Hours Viewed'] = df['Hours Viewed'].str.replace(',', '').astype(int)
df["season"] = df["season"].astype(int)
```

```python
season_counts = df['new_title'].value_counts()

# Find titles with only one season
titles_with_one_season = season_counts[season_counts == 1].index

# Filter the DataFrame to exclude rows with titles that have only one season
df = df[~df['new_title'].isin(titles_with_one_season)]
df
```

```python
# Group by 'new_title' and plot each group separately
fig, ax = plt.subplots(figsize=(13, 10))

for title, group in df.groupby('new_title'):
    ax.plot(group['season'], group['Hours Viewed'], marker='o', label=title)

ax.set_xlabel('Season')
ax.set_ylabel('Hours Viewed (log scale)')
ax.set_title('Hours Viewed per Season of Reference Class Dating TV Shows (Log Scale)')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_yscale('log')  # Set logarithmic scale for the y-axis
plt.show()
```
