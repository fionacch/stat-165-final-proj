```python
# Load the dataset that only shows the popular love_reality shows from Netflix
love_reality_df = pd.read_csv("/content/drive/MyDrive/Stat165_Final/love_reality.csv")

# Display the first few rows of the dataframe
love_reality_df.head()
```

```python
from datetime import datetime

# Convert 'Hours Viewed' to numerical data type by removing commas and converting to integer
love_reality_df['Hours Viewed'] = love_reality_df['Hours Viewed'].str.replace(',', '').astype(int)

# Convert 'Release Date' to datetime and then extract year and month as separate features
love_reality_df['Release Date'] = pd.to_datetime(love_reality_df['Release Date'])
love_reality_df['Release Year'] = love_reality_df['Release Date'].dt.year
love_reality_df['Release Month'] = love_reality_df['Release Date'].dt.month

# Since we're interested in the trend of Single's Inferno, let's isolate that series
singles_inferno_df = love_reality_df[love_reality_df['Title'].str.contains("Single’s Inferno")]
```

```python
# Calculate the growth rate in hours viewed from season 1 to season 2 for Single's Inferno
singles_inferno_df = singles_inferno_df.sort_values(by='Release Date')
singles_inferno_growth_rate = (singles_inferno_df['Hours Viewed'].iloc[-1] / singles_inferno_df['Hours Viewed'].iloc[-2]) - 1

# Add a new feature for 'Single's Inferno' that indicates the growth rate
love_reality_df['Singles Inferno Growth Rate'] = singles_inferno_growth_rate

# Since we know that Season 4 will be released in December, we can add that as well
# Let's assume it wll be mid-December similar to previous seasons
predicted_release_month = 12  # December
predicted_release_year = 2024  # We're predicting for Season 4, which we assume is in 2024

# Show the updated dataframe and the calculated growth rate
love_reality_df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Prepare the dataset for modeling
# Convert 'Available Globally?' to numeric (binary) data
love_reality_df['Available Globally?'] = love_reality_df['Available Globally?'].map({'Yes': 1, 'No': 0})

# We'll exclude the 'Unnamed: 0' and 'Title' columns as they are not relevant for modeling
model_df = love_reality_df.drop(columns=['Unnamed: 0', 'Title', 'Release Date'])

# Since the increase rate might not be linear, we can transform our target variable to use a logarithmic scale
# We add a small constant to avoid taking the log of zero
model_df['Log Hours Viewed'] = np.log(model_df['Hours Viewed'] + 1)
```

```python
model_df = model_df.dropna()
```

```python
model_df.head(10)
```

```python
# Model Selection
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

X = model_df[['Available Globally?', 'Release Year', 'Release Month', 'Singles Inferno Growth Rate']]
y = model_df[['Hours Viewed', 'Log Hours Viewed']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Model Training
model = RandomForestRegressor() # or use LinearRegression() for a simpler model
model.fit(X_train, y_train)
```

```python
# Model Evaluation
y_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f'RMSE: {rmse}')
```

```python
# feature engineering

f_netflix_global['Year'] = f_netflix_global['Release Date'].dt.year
f_netflix_global['Month'] = f_netflix_global['Release Date'].dt.month
f_netflix_global['Available Globally?'] = f_netflix_global['Available Globally?'].map({'Yes': 1, 'No': 0})
current_date = pd.to_datetime('2024-04-14')
f_netflix_global['Days Since Release'] = (current_date - f_netflix_global['Release Date']).dt.days
```

```python
# Model Training
model = RandomForestRegressor() # or use LinearRegression() for a simpler model
model.fit(X_train, y_train)
```

```python
# Model Evaluation
y_pred = model.predict(X_test)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f'RMSE: {rmse}')
```
